```python
# Read the data from the CSV file with read from pandas
data = pd.read_csv("Salary_dataset.csv")


# Configure the style
sns.set_theme(  palette="light:m_r")

# Plot the data points with seaborn

sns.jointplot(data=data, x="YearsExperience", y="Salary")
```

```python
# We're going to use this library and find out how it works on the Internet
from sklearn.linear_model import LinearRegression

# Separate the features (YearsExperience) from the target variable (Salary)
x = data[["YearsExperience"]]
y = data["Salary"]

# Create a linear regression model
regression = LinearRegression()


# Train the model
regression.fit(x.values,y)

# Predict the Salary values based on YearsExperience
predictions = regression.predict(x.values)
```

```python
# Create a linear regression plot using lmplot
sns.set_theme()
ax = sns.lmplot(x="YearsExperience", y="Salary", data=data)

# Set labels and title
ax.set(xlabel='Years of Experience', title='Linear Regression: Years of Experience vs Salary')
```
